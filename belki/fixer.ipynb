```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv('iris_2_classes.csv', sep=',')
```

```python
dataset
```

```python
for i in dataset.columns[:-1]:
    diff = dataset[i].max() - dataset[i].min()
    dataset[i] = dataset[i] + (diff * np.random.random(dataset.shape[0]) - diff)
```

```python
plt.scatter(
    dataset[dataset['variety'] == 'Setosa']['sepal.length'],
    dataset[dataset['variety'] == 'Setosa']['sepal.width']
)
plt.scatter(
    dataset[dataset['variety'] == 'Versicolor']['sepal.length'],
    dataset[dataset['variety'] == 'Versicolor']['sepal.width']
)
plt.show()
```

```python
plt.scatter(
    dataset[dataset['variety'] == 'Setosa']['petal.length'],
    dataset[dataset['variety'] == 'Setosa']['petal.width']
)
plt.scatter(
    dataset[dataset['variety'] == 'Versicolor']['petal.length'],
    dataset[dataset['variety'] == 'Versicolor']['petal.width']
)
plt.show()
```

```python
dataset.to_csv('iris_2_classes_fixed.csv', sep=',', index=False)
```
